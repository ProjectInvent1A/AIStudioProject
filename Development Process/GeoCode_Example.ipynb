```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from geopy import distance
```

```python
#gathers row's city and state
#from the video use Rate Limiter to perform bulk operations (not sure about this one yet)
#geolocator = RateLimiter(Nominatim(user_agent="Community_Matching").geocode)
geolocator = Nominatim(user_agent="Community_Matching")
city_state = df.iloc[10]['city'] + ", " + df.iloc[10]['state']
print(city_state)
```

```python
#finds latitude and longitude of the city and state
location = geolocator.geocode(city_state)
print(city_state, " - ", "latitude: ", location.latitude, " longitude: ", location.longitude)
```

```python
#example with zip code
zip_code = geolocator.geocode(90815)
print(zip_code, " - ", "latitude: ", zip_code.latitude, " longitude: ", zip_code.longitude)
```

```python
#calculating distance
starting_point = (zip_code.latitude, zip_code.longitude)
ending_point = (location.latitude, location.longitude)

distance.distance(starting_point, ending_point).miles

#should determine cutoff of what is too far
```

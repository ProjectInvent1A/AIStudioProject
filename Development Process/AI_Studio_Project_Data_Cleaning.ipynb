Now using the Charity Navigator API to add info to dataset

```python
# function that will call the API given a string name
def callAPI(name):
  url = 'https://data.charitynavigator.org/graphql'
  query = """
  query PublicSearchFaceted($term: String!) {
      publicSearchFaceted(term: $term) {
        results {
          mission
          organization_url
          cause
        }
      }
    }
  """

  variables = {
      'term': name
  }

  payload = {
      'query': query,
      'variables': variables
  }

  headers = {
      'Stellate-Api-Token' : 'stl8_18009830b2a4f44b8fba693508439c4fcf06850aa8034e98f34821f442684360',
      'Content-Type' : 'application/json',
      'User-Agent' : 'PostmanRuntime/7.37.3',
      'Accept' : '*/*',
      'Connection' : 'keep-alive'
  }

  response = requests.post(url, json=payload, headers=headers)

  if(response.status_code != 200):
    print("response status code: ", response.status_code)

  api_response_dict = json.loads(response.content)
  return api_response_dict
```

```python
# test data frame to try this on
df_test = pd.concat([df_fullInfo.tail(10)])
df_test
```

```python
# make api calls for each element in the dataframe
dict_list = []
for name in df_test["name"]:
  response = callAPI(name)
  print(response)
  if(response["data"] == None):
    api_dict = {'mission': None,'organization_url': None,'cause': None, 'name': name}
  else:
    api_dict = response['data']['publicSearchFaceted']['results'];
    if(len(api_dict)) != 0:
      api_dict = api_dict[0]
      api_dict['name'] = name
    else:
      api_dict = {'mission': None,'organization_url': None,'cause': None, 'name': name}
  dict_list.append(api_dict)
```

```python
# create a new dataframe with the returned info
api_call_df = pd.DataFrame(dict_list)
api_call_df
```

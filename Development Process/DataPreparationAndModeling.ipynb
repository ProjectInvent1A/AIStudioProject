```python
#combining text data about organization's purpose into one new column, Text
df["mission"] = df["mission"].fillna("")
df["cause"] = df["cause"].fillna("")
df["Title"] = df["Title"].fillna("")
df["Description"] = df["Description"].fillna("")

df['Text'] = "";
df['Text'] += np.where(df["cause"] != "", df["cause"] + ". ", "")
df['Text'] += np.where(df["mission"] != "", df['mission'] + ". ", "")
df['Text'] += np.where(df["Title"] != "", df['Title'] + ". ", "")
df['Text'] += np.where(df["Description"] != "", df['Description'] + ". ", "")
```

Dropping duplicates in ein identification number so the same organizations aren't in here multiple times

```python
duplicates_ein = df[df.duplicated(subset='org_ein', keep=False)]
duplicates_ein
```

```python
duplicates_ein_drop = df[df.duplicated(subset='org_ein', keep="first")]
duplicates_ein_drop
```

```python
df = df.drop_duplicates(subset='org_ein', keep="first")
df.shape
```

```python
df.isna().sum()
```

```python
# counting how many rows still have empty or unhelpful values
print((df['Text'] == "").sum())
print((df["Text"] == "Unknown or not classified. ").sum())
```

```python
# dropping rows with empty and unhelpful values
df = df[df['Text'] != ""]
df = df[df["Text"] != "Unknown or not classified. "]
df.shape
```

```python
# dropping rows that are missing city, state, or ein
df.dropna(subset=['org_ein', 'city', 'state'], inplace=True)
```
